# DATA ENGINEERING

### Data Cleaning

```python
# Convert Price
df['price'] = (
    df['price']
    .replace('[\$,]', '', regex=True)
    .astype(float)
)
```

```python
# Convert Date Columns
date_columns = [
    'host_since',
    'last_scraped',
    'calendar_last_scraped',
    'first_review',
    'last_review'
]

for col in date_columns:
    df[col] = pd.to_datetime(df[col], errors='coerce')
```

```python
# Verify Dates
df[date_columns].info()
```

```python
# Check Negative Prices
(df['price'] < 0).sum()
```

```python
# Latitude & Longitude Validation
df[['latitude','longitude']].describe()
```

# HOST ANALYSIS

```python
# Number of Listings per Host
host_listings = df['host_id'].value_counts()

host_listings.describe()
```

```python
# Top 10 Hosts with Most Listings
top_hosts = df['host_id'].value_counts().head(10)

import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
top_hosts.plot(kind='bar')
plt.title("Top 10 Hosts by Number of Listings")
plt.xlabel("Host ID")
plt.ylabel("Number of Listings")
plt.xticks(rotation=45)
plt.show()
```

```python
# Superhost vs Non-Superhost Analysis
df['host_is_superhost'] = df['host_is_superhost'].map({'t': 1, 'f': 0})

df.groupby('host_is_superhost')['price'].mean()
```

```python
# Visualization
df.groupby('host_is_superhost')['price'].mean().plot(kind='bar')
plt.title("Average Price: Superhost vs Non-Superhost")
plt.show()
```
